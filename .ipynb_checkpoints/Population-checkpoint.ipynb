```python
def accent_replace(df):
    df_cat = df.select_dtypes(['object'])
    for i in df_cat.columns:
        df[i]=df[i].str.replace('á', 'a')
        df[i]=df[i].str.replace('é', 'e')
        df[i]=df[i].str.replace('í', 'i')
        df[i]=df[i].str.replace('ó', 'o')
        df[i]=df[i].str.replace('ú', 'u')
        df[i]=df[i].str.replace('ñ', 'ni')

    return df
```

```python
def items_lower(df):
    df_cat = df.select_dtypes(['object'])
    for i in df_cat.columns:
        df[i] = df[i].str.lower()
        
    return df
```

# Create query from SQL

```python
import pymysql
from sqlalchemy import create_engine

import getpass  # To get the password without showing the input
password = getpass.getpass()
```

```python
# get the data
connection_string = 'mysql+pymysql://root:' + password + '@localhost/population'
engine = create_engine(connection_string)
query = '''SELECT pob.communities, pob.year, emp.total_number_of_companies, emp.total_capital_subscribed, emp.total_capital_paid_up,
ipv.house_price_index_hpi, ipc.consumer_price_index_cpi, mor.cost_per_mortgage, birth.total_births, birth.total_marriages, 
birth.total_functions, occupied.occupied_population, gdp.total_gdp, pob.total_population
FROM population.poblacion_por_comunidades pob
JOIN population.ipv_por_comunidad ipv 
ON pob.communities = ipv.communities AND pob.year = ipv.year
JOIN population.ipc_por_comunidad ipc 
ON pob.communities = ipc.communities AND pob.year = ipc.year
JOIN population.nacimiento_matrimonio_defuncion_comunidades birth
ON pob.communities = birth.communities AND pob.year = birth.year
JOIN population.numero_de_ocupados_por_comunidad_autonoma occupied
ON pob.communities = occupied.communities AND pob.year = occupied.year
JOIN population.numero_hipotecas_importe_hipotecas_por_comunidad_dividir mor 
ON pob.communities = mor.communities AND pob.year = mor.year
JOIN population.PIB_Valor_ComunidadesAutonomas gdp
ON pob.communities = gdp.communities AND pob.year = gdp.year
JOIN population.empresas_y_capital_por_comunidad emp
ON pob.communities = emp.communities AND pob.year = emp.year
WHERE pob.year IN (2020,2019,2018)
ORDER BY communities;'''


df_sql_1 = pd.read_sql_query(query, engine)
df_sql_1.head()
```

# Regression model

```python
#Import sklearn train_test_split and separate the data.
```

```python
X = df_sql_3.drop(['communities', 'year', 'total_population','total_number_of_companies','occupied_population','total_gdp','total_births','total_functions','total_marriages'],axis=1)
y = df_sql_3['total_population']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=100)
```

```python
# Normalize the continuous variables.
```

```python
numericals_train = X_train.select_dtypes(np.number)
numericals_test = X_test.select_dtypes(np.number)
numericals_train.head()
```

```python
from sklearn.preprocessing import StandardScaler

transformer = StandardScaler().fit(numericals_train)
numericals_train_standardized = transformer.transform(numericals_train)
numericals_test_standardized = transformer.transform(numericals_test)
```

```python
X_train = numericals_train_standardized
X_test = numericals_test_standardized
```

```python
# Function that takes a list of models and train and tests them
```

```python
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.linear_model import LinearRegression
```

```python
def model(X_train, X_test, y_train, y_test):
    
    LR = LinearRegression()
    KNN = KNeighborsRegressor(n_neighbors=6)
    
    model_list = [LR, KNN]
    
    for i in model_list:
        
        i.fit(X_train,y_train)
        score = i.score(X_test, y_test)
        print(i, 'model:', score)
```

```python
model(X_train, X_test, y_train, y_test)
```

```python
# The regressions after dropping 'communities', 'year', 'total_population','total_number_of_companies','occupied_population','total_gdp','total_births','total_functions','total_marriages'
# are 0.77 for linear and 0.82 for kneighbors. 
```
